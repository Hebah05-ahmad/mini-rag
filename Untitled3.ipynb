```python
base_controller_content = '''from helpers.config import get_settings
import os

class BaseController:
    def __init__(self):
        self.app_settings = get_settings()
        self.base_path = os.path.dirname(os.path.dirname(__file__))
        self.files_path = os.path.join(self.base_path, "assets/files")
        os.makedirs(self.files_path, exist_ok=True)
'''

with open("source/controllers/base_controller.py", "w", encoding="utf-8") as f:
    f.write(base_controller_content)
```

```python
project_controller_content = '''from controllers.base_controller import BaseController
import os

class ProjectController(BaseController):
    def get_project_path(self, project_id: str):
        project_dir = os.path.join(self.files_path, project_id)
        if not os.path.exists(project_dir):
            os.makedirs(project_dir)
        return project_dir

    def list_projects(self):
        projects = []
        if os.path.exists(self.files_path):
            projects = [d for d in os.listdir(self.files_path)
                       if os.path.isdir(os.path.join(self.files_path, d))]
        return projects
'''

with open("source/controllers/project_controller.py", "w", encoding="utf-8") as f:
    f.write(project_controller_content)
```

```python
data_controller_content = '''from controllers.base_controller import BaseController
from models.enums.response_enums import ResponseSignal
import random
import string
import re

class DataController(BaseController):
    def validate_file(self, file):
        if file.content_type not in self.app_settings.FILE_ALLOWED_TYPES:
            return False, ResponseSignal.FILE_TYPE_NOT_SUPPORTED.value
        if file.size > (self.app_settings.FILE_MAX_SIZE * 1048576):
            return False, ResponseSignal.FILE_SIZE_EXCEEDED.value
        return True, None

    def generate_unique_name(self, filename: str):
        random_str = ''.join(random.choices(string.ascii_letters + string.digits, k=12))
        clean_name = re.sub(r'[^\w\.]', '_', filename)
        return f"{random_str}_{clean_name}"

    def get_file_info(self, file_path: str):
        import os
        stats = os.stat(file_path)
        return {
            "size_kb": stats.st_size / 1024,
            "created": stats.st_ctime,
            "modified": stats.st_mtime,
            "extension": os.path.splitext(file_path)[1]
        }
'''

with open("source/controllers/data_controller.py", "w", encoding="utf-8") as f:
    f.write(data_controller_content)
```

```python
data_routes_content = '''from fastapi import APIRouter, UploadFile, status, BackgroundTasks
from fastapi.responses import JSONResponse
from controllers.data_controller import DataController
from controllers.project_controller import ProjectController
from models.enums.response_enums import ResponseSignal
import aiofiles
import os
import asyncio

data_router = APIRouter(prefix="/api/v1/data", tags=["Data"])

async def save_file_chunks(file, file_path, chunk_size):
    async with aiofiles.open(file_path, 'wb') as f:
        while True:
            chunk = await file.read(chunk_size)
            if not chunk:
                break
            await f.write(chunk)

@data_router.post("/upload/{project_id}")
async def upload_file(project_id: str, file: UploadFile, background_tasks: BackgroundTasks):
    data_ctrl = DataController()
    is_valid, error_signal = data_ctrl.validate_file(file)

    if not is_valid:
        return JSONResponse(
            status_code=status.HTTP_400_BAD_REQUEST,
            content={"signal": error_signal}
        )

    try:
        proj_ctrl = ProjectController()
        project_path = proj_ctrl.get_project_path(project_id)
        unique_name = data_ctrl.generate_unique_name(file.filename)
        file_path = os.path.join(project_path, unique_name)

        await save_file_chunks(file, file_path, data_ctrl.app_settings.FILE_DEFAULT_CHUNK_SIZE)

        return {
            "signal": ResponseSignal.FILE_UPLOAD_SUCCESS.value,
            "file_name": unique_name,
            "original_name": file.filename,
            "project_id": project_id,
            "size_bytes": file.size
        }

    except Exception as e:
        return JSONResponse(
            status_code=status.HTTP_500_INTERNAL_SERVER_ERROR,
            content={
                "signal": ResponseSignal.FILE_UPLOAD_FAILED.value,
                "error": str(e)
            }
        )

@data_router.get("/projects")
async def list_all_projects():
    proj_ctrl = ProjectController()
    projects = proj_ctrl.list_projects()
    return {"projects": projects, "count": len(projects)}

@data_router.get("/project/{project_id}/files")
async def list_project_files(project_id: str):
    proj_ctrl = ProjectController()
    project_path = proj_ctrl.get_project_path(project_id)

    files = []
    if os.path.exists(project_path):
        files = os.listdir(project_path)

    return {"project_id": project_id, "files": files, "count": len(files)}
'''

with open("source/routes/data.py", "w", encoding="utf-8") as f:
    f.write(data_routes_content)
```
